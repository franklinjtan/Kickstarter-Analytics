# Importing necessary libaries and packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
# Load the CSV dataset
df = pd.read_csv('../inputs/kickstarter_projects.csv')
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Check the summary statistics of numerical columns
print(df.describe())
```

```python
# Check the data types and missing values
print(df.info())
```

# Plot the distribution of project states

```python
state_counts = df['state'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(state_counts.index, state_counts.values)
plt.xlabel('Project State')
plt.ylabel('Count')
plt.title('Distribution of Project States')
plt.show()
```

# Plot the distribution of project categories

```python
category_counts = df['category'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Top 10 Project Categories')
plt.xticks(rotation=45)
plt.show()
```

# Plot the success rate by main category

```python
success_rate_by_category = df.groupby('main_category')['state'].apply(lambda x: (x == 'successful').mean())
plt.figure(figsize=(10, 6))
plt.bar(success_rate_by_category.index, success_rate_by_category.values)
plt.xlabel('Main Category')
plt.ylabel('Success Rate')
plt.title('Success Rate by Main Category')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()
```

# Plot the success rate by country

```python
success_rate_by_country = df.groupby('country')['state'].apply(lambda x: (x == 'successful').mean())

# Sort the countries by success rate in descending order
success_rate_by_country = success_rate_by_country.sort_values(ascending=False)

# Plot the success rate by country
plt.figure(figsize=(12, 6))
plt.bar(success_rate_by_country.index, success_rate_by_country.values)
plt.xlabel('Country')
plt.ylabel('Success Rate')
plt.title('Success Rate by Country')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()
```

# Number of successful projects by year

```python
df['launched'] = pd.to_datetime(df['launched'])
df['year'] = df['launched'].dt.year
success_counts_by_year = df[df['state'] == 'successful']['year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(success_counts_by_year.index, success_counts_by_year.values)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Number of Successful Projects by Year')
plt.show()
```

# Scatter plot of pledged amount vs. funding goal:

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['usd_goal_real'], df['usd_pledged_real'], alpha=0.5)
plt.xlabel('Funding Goal (USD)')
plt.ylabel('Pledged Amount (USD)')
plt.title('Pledged Amount vs. Funding Goal')
plt.show()
```
