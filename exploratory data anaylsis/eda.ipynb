# Plot the success rate by main category

```python
success_rate_by_category = df.groupby('main_category')['state'].apply(lambda x: (x == 'successful').mean())
plt.figure(figsize=(10, 6))
plt.bar(success_rate_by_category.index, success_rate_by_category.values)
plt.xlabel('Main Category')
plt.ylabel('Success Rate')
plt.title('Success Rate by Main Category')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()
```

# Plot the success rate by country

```python
success_rate_by_country = df.groupby('country')['state'].apply(lambda x: (x == 'successful').mean())

# Sort the countries by success rate in descending order
success_rate_by_country = success_rate_by_country.sort_values(ascending=False)

# Plot the success rate by country
plt.figure(figsize=(12, 6))
plt.bar(success_rate_by_country.index, success_rate_by_country.values)
plt.xlabel('Country')
plt.ylabel('Success Rate')
plt.title('Success Rate by Country')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()
```

```python
# Sort the DataFrame by the "backers" column in descending order
sorted_df = df.sort_values(by='backers', ascending=False)

# Get the top 5 "name" values with the highest number of backers
top_5_backers = sorted_df.head(5)[['name', 'backers']]

# Convert the top 5 "name" values into a DataFrame
top_5_backers_df = pd.DataFrame(top_5_backers)

# Print the top 5 "name" DataFrame
top_5_backers_df
```

```python
from pytrends.request import TrendReq
import time

# Create pytrends object
pytrends = TrendReq()

# Set the timeframe for the past 5 years
timeframe = 'today 5-y'

# Iterate through the top 5 names and retrieve Google Trends data
for name in top_5_backers_df['name']:
    pytrends.build_payload(kw_list=[name], timeframe=timeframe)
    interest_over_time = pytrends.interest_over_time()
    print(f"Google Trends analysis for {name} (Past 5 Years):\n")
    with pd.option_context('display.max_rows', None):  # Set the option to display all rows
        print(interest_over_time)
    print("\n")
    time.sleep(60)  # Add a delay of 60 seconds between each request
```
